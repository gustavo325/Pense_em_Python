### Exercício 2.2 - Questão 1

##### O volume de uma esfera com raio r é 4/3 * pi * r³ – Volume de uma esfera.. Qual é o volume de uma esfera com raio 5?

```python
r = 5
PI = 3.14
volume = 4/3 * PI *r**3
print(f'O volume de uma esfera com raio 5 é igual a {volume:.2f}')
```

### Exercício 2.2 - Questão 2

##### Suponha que o preço de capa de um livro seja 24,95 reais, mas as livrarias recebem um desconto de 40%. O transporte custa 3,00 reais para o primeiro exemplar e 75 centavos para cada exemplar adicional. Qual é o custo total de atacado para 60 cópias?

```python
capa_livro = 24.95
transporte_primeiro_livro = 3
transporte_exemplar_adicional = 0.75
desconto = capa_livro * 0.4

# Custo para 60 cópias: 

qtde = 60

custo = (capa_livro - desconto) * qtde + transporte_primeiro_livro + (transporte_exemplar_adicional * (qtde-1))
print(f'{custo:.2f}')
```

### Exercício 2.2 - Questão 3

##### Se eu sair da minha casa às 6:52 e correr 1 quilômetro a um certo passo (8min15s por quilômetro), então 3 quilômetros a um passo mais rápido (7min12s por quilômetro) e 1 quilômetro no mesmo passo usado em primeiro lugar, que horas chego em casa para o café da manhã?

```python
# tempo gasto em segundos

tempo_inicial_em_segundos = 24720 
primeiro_quilometro = 495
tres_quilometros_seguintes = 432

tempo_chegada = tempo_inicial_em_segundos + (primeiro_quilometro*2) + (tres_quilometros_seguintes*3)

# convertendo em horas

tempo_chegada /= 60 
tempo_chegada /= 60
hora = str(tempo_chegada)[0]
hora = int(hora)
minuto = str(tempo_chegada)[2:]
minuto = '0.' + minuto
minuto = float(minuto)*60
minuto = int(minuto)

print(f'O horário de chegada é: {hora} Hrs e {minuto} minutos')
```
