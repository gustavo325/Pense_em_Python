#### Escreva um script que leia a hora atual e a converta em um tempo em horas, minutos e segundos, mais o número de dias desde a época.

```python
import time

def converte_tempo(segundos):
  minutos = segundos //60
  hora = 0
  dia = 0
  while minutos >= 60: 
    hora += 1
    minutos -= 60
  while hora >= 24: 
    dia += 1
    hora -= 24
  resto_segundos = (segundos - (segundos //60)*60)

  print(f'Agora são {hora-3} {"horas" if hora > 1 else "hora"}, {minutos:.0f} "\
  "{"minutos" if minutos > 1 else "minuto"} e {resto_segundos:.0f} {"segundos." if resto_segundos >1 else "segundo."}')
  print(f'\nA quantidade de dias passados desde 1970 é de {dia} dias.')

converte_tempo(time.time())
```

O último teorema de Fermat diz que não existem números inteiros a, b e c tais que a**n + b**n == c**n para quaisquer valores de n maiores que 2.

Escreva uma função chamada check_fermat que receba quatro parâmetros – a, b, c e n – e verifique se o teorema de Fermat se mantém. Se n for maior que 2 e a**n + b**n == c**n o programa deve imprimir, “Holy smokes, Fermat was wrong!” Senão o programa deve exibir “No, that doesn’t work.”

Escreva uma função que peça ao usuário para digitar valores para a, b, c e n, os converta em números inteiros e use check_fermat para verificar se violam o teorema de Fermat.

```python
# Teorema de Fermat 

def check_fermat(a,b,c,n): 
  if n > 2 and a**n + b**n == c**n: 
    print("Holy Smokes, Fermat was wrong!")
  else: 
    print("No, that doesn't work.")

def valores():
  a = int(input("Digite o valor de a do teorema de Fermat"))
  b = int(input("Digite o valor de b do teorema de Fermat"))
  c = int(input("Digite o valor de c do teorema de Fermat"))
  n = int(input("Digite o valor de n do teorema de Fermat"))
  check_fermat(a,b,c,n)

valores()
```

Há um teste simples para ver se é possível formar um triângulo para quaisquer três comprimentos:

##### Se algum dos três comprimentos for maior que a soma dos outros dois, então você não pode formar um triângulo. Senão, você pode. (Se a soma de dois comprimentos igualar o terceiro, eles formam um triângulo chamado “degenerado”)

##### 1. Escreva uma função chamada is_triangle que receba três números inteiros como argumentos, e que imprima “Yes” ou “No”, dependendo da possibilidade de formar ou não um triângulo de gravetos com os comprimentos dados.

```python
def is_triangle(x1, x2, x3): 
  if (x1 < x2+x3) and x2 < x1+x3 and x3 <x1+x2:
    print('Yes')
  else: 
    print('No')

def comprimentos(): 
  x1 = int(input("Digite o comprimento do 1º graveto"))
  x2 = int(input("Digite o comprimento do 2º graveto"))
  x3 = int(input("Digite o comprimento do 3º graveto"))
  is_triangle(x1, x2, x3) 

comprimentos()
```
